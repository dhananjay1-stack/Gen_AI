```python
import random
import textwrap
```

```python
text = """
The quick brown fox jumps over the lazy dog.
The quick brown fox is very fast.
The lazy dog sleeps all day long.
The quick brown fox wakes up the lazy dog.
"""
```

```python
def build_markov_chain(data):
    # Split text into words based on whitespace
    words = data.split()

    # Initialize a dictionary to hold our model
    # Format: { 'current_word': ['next_word_option1', 'next_word_option2'] }
    chain = {}

    # Iterate through the words (stopping before the last one)
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        # If the word isn't in our dictionary yet, add it
        if current_word not in chain:
            chain[current_word] = []

        # Append the next word to the list of possibilities
        chain[current_word].append(next_word)

    return chain
```

```python
def generate_text(chain, start_word, length=20):
    # If the start word isn't in our text, pick a random one
    current_word = start_word if start_word in chain else random.choice(list(chain.keys()))

    result = [current_word]

    for _ in range(length):
        # If the current word is at the end of the text and has no followers, stop
        if current_word not in chain:
            break

        # Pick a random next word from the list of possibilities
        next_word_options = chain[current_word]
        next_word = random.choice(next_word_options)

        # Add to results and update current_word
        result.append(next_word)
        current_word = next_word

    return " ".join(result)
```

```python
# Build the model
markov_model = build_markov_chain(text)

# Generate text
# We try to start with "The", but you can change this.
generated_output = generate_text(markov_model, start_word="The", length=15)

print("--- Generated Text ---")
print(textwrap.fill(generated_output, width=50))
```
